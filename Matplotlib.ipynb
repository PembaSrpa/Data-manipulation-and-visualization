# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

## Line plot

### What is a line plot?
A continuous line that connects data points.
used for:
* Trends over time(days, weeks, months, years)
* Comparing multiple trends
* Seeing how something changes over time

Best when your x-axis is time or another continuous variable

Avoid when you have categorical data.

```python
x = np.arange(1,10,0.1)
y = 2*x+5
```

```python
plt.plot(x,y)
plt.show()
```

## Customizing line plot

```python
x = np.arange(1,10,0.1)
y = 2*x+5
plt.plot(x,y, color='r')
plt.title('Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Adding Two subplots in the same graph

```python
x = np.arange(1,10,0.1)
y1 = 2*x+5
y2 = 3*x+10

plt.subplot(1, 2, 1)
plt.plot(x,y1)
plt.subplot(1, 2, 2)
plt.plot(x,y2)

plt.show()
```

## Bar-plot

### What is a Bar-plot?
Rectangular bars that represents values.
Used for:
* Comparing categories (e.g., sales by product)
* Highlighting highest/lowest groups.
* Showing counts or percentages.

Best when your x-axis is categorical or discrete.

Avoid when you have continuous data.

```python
# Dictionary
fruit = {'apple':30, 'mango':45, 'banana':20, 'pineapple':5}
names = list(fruit.keys())
quantity = list(fruit.values())
```

```python
names, quantity
```

```python
plt.bar(names, quantity)
plt.show
```

## Customizing the Bar plot

```python
plt.bar(names, quantity, color='orange')
plt.title('Distribution of fruits')
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show
```

## Making Bar Horizontal

```python
plt.barh(names, quantity, color='green')
plt.title('Distribution of fruits')
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show
```

## Scatter-plot

### What is a Scatter-plot?
Points plotted on an x-y coordinate system.
Used for:
* Showing relationships/correlation between two variables.
* Detecting clusters, trends, and outliers.

Best when both X and Y are numeric.

Avoid when you have categories only.

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
a = [1, 2, 3, 4, 5, 4, 3, 2, 9]
b = [9, 8, 7, 6, 5, 4, 3, 2, 1]

plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

## To change the size of the points

```python
# Changes the figure size
plt.figure(figsize=(10,10))
plt.scatter(x,a, s = 200)
# Marker changes shapes
plt.scatter(x,b, s = 500, marker= '2')
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

## Histogram

### What is a Histogram?
Bar-like chart but grouped into bins.
Used for:
* Showing distribution of a single numeric variable.
* Understanding spread: skewness, peaks, tails.

Best when you want to see how often values occur within ranges.

Avoid when you are comparing many groups.

```python
data = [1, 2, 3, 3, 4, 5, 6, 8, 1, 2, 3, 1, 1, 4]
plt.hist(data)
plt.show()
```

## Box plot

### What is a Box plot?
Shows quartiles (min, quartiles, max) and outliers.
Used for:
* Comparing several groups.
* Spotting spread, medians, and outliers quickly.

Best when you don't need the exact shape of the distribution; you just want the stats.

Avoid when showing distribution shape.

```python
one = [1, 2, 3, 4, 5, 6, 7, 8, 9]
two = [1, 2, 3, 4, 5, 4, 3, 2, 1]
three = [9 ,8, 7, 6, 6, 7, 8, 9, 6]

data = list([one, two, three])

plt.boxplot(data)
plt.show()
```

## Violin plot

### What is a Violin plot?
A symmetrical density plot around a central axis, showing the distribution of data.
Used for:
* Showing distribution shape and summary statistics together.
* Comparing distributions of different groups.

Best when you need to understand the full distribution (not just quartiles)

Avoid when audience is non-technical.

```python
one = [1, 2, 3, 4, 5, 6, 7, 8, 9]
two = [1, 2, 3, 4, 5, 4, 3, 2, 1]
three = [9 ,8, 7, 6, 6, 7, 8, 9, 6]

data = list([one, two, three])

plt.violinplot(data)
# we can also add grid to our plot
plt.grid(True)
plt.title('Distribution of data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# we can also add median in violin plot
plt.violinplot(data, showmedians=True)
# just like that we can show means using show means=True
plt.show()
```

## Pie Chart

### What is a Pie Chart?
A circular chart divided into slices to show proportions.
Used for:
* Showing parts of a whole.
* Comparing proportions of different categories.

Best when you have 4-6 categories.

Avoid when you have more than 6 categories.

```python
fruit = ['apple', 'orange', 'banana', 'grapes']
quant = [12,23,45,56]
plt.pie(quant, labels=fruit)
plt.show()
```

```python
# to add percentage: autopct='%0.1f%%'
# to add shadow: shadow=True
# to highlight/separate one part: explode=()
plt.pie(quant, labels=fruit, autopct='%0.1f%%', shadow=True, explode=(0,0,0,0.1))
plt.show()
```

```python
# Donut plot
pie1 = plt.pie(quant, labels=fruit, autopct='%0.1f%%', radius=2)
pie2 = plt.pie([5], colors='w', radius=1)
plt.show()
```

## Area plot

### What is an Area plot?
A line plot where the area under the line is filled with color.
Used for:
* Showing cumulative trends over time.
* Visualizing how different variables contribute to a total.
* Making trends look more visually "weighted" or emphasized.


Best when you want to highlight volume, magnitude or total changes over time.

Avoid when too many series are plots.

```python
x = [10,20,30,40,50,60,70,80,90]
y = [1,1,2,5,4,6,7,3,8]

plt.stackplot(x,y)
plt.show()
```
