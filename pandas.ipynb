# Introduction to Pandas

## What is Pandas?
A powerful library for data manipulation and analysis in Python.
Built on top of NumPy, provides data structures like Series and DataFrame.

## Why use Pandas?
* Easy data cleaning and transformation.
* Efficient handling of large datasets.
* Integration with other Python libraries like Matplotlib and Seaborn for visualization.

## To Merge

```python
df6 = pd.DataFrame({
    "First Name": ["Harry","WTF","Ashutosh","ahsg"],
    "Marks": [12,23,34,23]
})
```

```python
df7 = pd.DataFrame({
    "First Name": ["Larry","F","asf","ag"],
    "Marks": [34,12,23,23]
})
```

```python
df6
```

```python
df7
```

```python
pd.concat([df6, df7])
```

```python
df8 = pd.DataFrame({
    "First Name": ["As","asd"],
    "Marks": [34,12]
})
```

```python
df8
```

```python
df9 = pd.DataFrame({
    "First Name": ["As","asd"],
    "Roll": [4,2]
})
```

```python
df9
```

```python
pd.merge(df8,df9, on="First Name")
```
